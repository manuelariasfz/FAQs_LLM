```python
import pandas as pd
import glob

pd.set_option('display.max_columns', None)

# Obtener la lista de todos los archivos CSV en el directorio actual que empiezan con "sessionlist-"
archivos_csv = glob.glob("sessionlist-*.csv")

# Crear una lista para almacenar los DataFrames de cada archivo CSV
dfs = []

# Iterar sobre cada archivo CSV y cargarlo en un DataFrame de Pandas
for archivo in archivos_csv:
    df = pd.read_csv(archivo)
    dfs.append(df)

# Concatenar todos los DataFrames en uno solo
df_final = pd.concat(dfs, ignore_index=True)
```

```python
import csv

# Convertir la columna "Message" en un diccionario
message_dict = df_final["Message"].to_dict()

# Escribir el diccionario en un archivo CSV
with open("messages.csv", "w", newline='', encoding='utf-8') as f:
    writer = csv.DictWriter(f, fieldnames=["Index", "Message"])
    writer.writeheader()
    for key, value in message_dict.items():
        writer.writerow({"Index": key, "Message": value})
```
